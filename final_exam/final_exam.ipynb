```python
import pandas as pd 
import numpy as np


# edit
def drop_duplicates(data):
    '''
    Xóa các dòng bị lặp
    - 1 button 
    '''
    return data.drop_duplicates()


def handle_missing_data(data):
    '''
    Hàm này xử lý các giá trị NaN
    Nếu số lượng NaN nhỏ hơn mức cho phép thì ta có thể drop
    Nếu không, ta thay giá trị mới cho NaN
    - button
    '''
    num_missing_rows = data.isnull().any(axis=1).sum()

    if num_missing_rows <= len(data) / 10:
        data = data.dropna()
    else:
        for col in data.columns:
            if data[col].dtype in ['int64', 'float64']:
                # Thay thế NaN bằng mean() cho giá trị định lượng
                data[col] = data[col].fillna(data[col].mean())
            else:
                # Thay thế NaN bằng ffill() cho giá trị định tính
                data[col] = data[col].fillna(method='ffill')
    return data


def adjust_data(data, lst=[], method='minmax'):
    """
    Hàm chuẩn hóa dữ liệu trong DataFrame
    đầu vào có data, các cột cần chuẩn hóa, phương pháp chuẩn hóa
    1 button chính - 2 checkbutton phụ
    """

    if lst == []:
        columns = data.columns
    else:
        columns = lst

    if method == 'minmax':
        # Chuẩn hóa min-max
        for col in columns:
            min_val = data[col].min()
            max_val = data[col].max()
            data[col] = (data[col] - min_val) / (max_val - min_val)
    elif method == 'standardize':
        # Chuẩn hóa Z-score
        for col in columns:
            mean_val = data[col].mean()
            std_val = data[col].std()
            data[col] = (data[col] - mean_val) / std_val
    elif method == 'normalize':
        # Chuẩn hóa L2 norm
        for col in columns:
            data[col] = data[col] / np.linalg.norm(data[col])
    return data





# transform
def change_data_type(data, col, type):
    """
    Hàm thay đổi kiểu dữ liệu của một cột trong Data
    1 button chính - 2 checkbutton phụ
    """
    data[col] = data[col].astype(type)
    return data


def merge_column(data, lst, char):
    """
    1 button chính, 1 checkbutton phụ, 1 entry
    Hàm hợp nhất các cột trong DataFrame thành một cột mới dựa trên giá trị char được nhập
    """

    new_col_name = lst[0] + char + lst[1]
    data[new_col_name] = data[lst].apply(lambda row: char.join(row.astype(str)), axis=1)
    return data


def split_column(data, col, char):
    """
    1 button chính - 1 checkbutton phụ, 1 entry
    Hàm tách một cột trong DataFrame thành nhiều cột mới
    """
    split_values = data[col].str.split(char, expand=True)
    new_col_names = [col + '_' + str(i + 1) for i in range(split_values.shape[1])]
    data[new_col_names] = split_values
    return data

# view
def entire_dataset(data):
    '''1 button'''
    return data

def specific_column(data, lst=[]):
    ''' 
    1 button chính, 1 checkbutton phụ
    '''
    return data[lst]
```

```python
import tkinter as tk
from tkinter import ttk

class InitialApp:
    def __init__(self, root):
        self.root = root
        self.setupUi()

    def setupUi(self):
        self.clear_widgets()
        
        self.label = ttk.Label(self.root, text="Initial Interface", font=("Arial", 24))
        self.label.pack(pady=20)
        
        self.button = ttk.Button(self.root, text="Go to Second Interface", command=self.go_to_second_interface)
        self.button.pack(pady=10)

    def go_to_second_interface(self):
        self.clear_widgets()
        SecondApp(self.root)
        
    def clear_widgets(self):
        for widget in self.root.winfo_children():
            widget.destroy()

class SecondApp:
    def __init__(self, root):
        self.root = root
        self.setupUi()

    def setupUi(self):
        self.clear_widgets()

        self.label = ttk.Label(self.root, text="Second Interface", font=("Arial", 24))
        self.label.pack(pady=20)
        
        self.button = ttk.Button(self.root, text="Back to Initial Interface", command=self.go_to_initial_interface)
        self.button.pack(pady=10)

    def go_to_initial_interface(self):
        self.clear_widgets()
        InitialApp(self.root)
        
    def clear_widgets(self):
        for widget in self.root.winfo_children():
            widget.destroy()

def main():
    root = tk.Tk()
    root.geometry("800x600")
    root.title("Tkinter App with Menu")
    
    def open_initial_interface():
        app = InitialApp(root)

    def open_second_interface():
        app = SecondApp(root)

    # Tạo menu
    menu = tk.Menu(root)
    root.config(menu=menu)
    
    # Tạo menu "Navigate"
    navigate_menu = tk.Menu(menu, tearoff=0)
    menu.add_cascade(label="Navigate", menu=navigate_menu)
    navigate_menu.add_command(label="Go to Initial Interface", command=open_initial_interface)
    navigate_menu.add_command(label="Go to Second Interface", command=open_second_interface)

    # Khởi tạo giao diện ban đầu
    app = InitialApp(root)

    root.mainloop()

if __name__ == "__main__":
    main()
```

```python
import pandas as pd

def frequency_distribution(df, column):
    """
    Trả về bảng phân phối tần số của một cột được chọn từ một DataFrame.

    Parameters:
    df (pd.DataFrame): DataFrame chứa dữ liệu.
    column (str): Tên cột cần tính phân phối tần số.

    Returns:
    pd.DataFrame: DataFrame chứa bảng phân phối tần số.
    """
    # Kiểm tra nếu cột tồn tại trong DataFrame
    if column not in df.columns:
        raise ValueError(f"Column '{column}' does not exist in the DataFrame")
    
    # Tính phân phối tần số
    frequency_df = df[column].value_counts().reset_index()
    frequency_df.columns = [column, 'Frequency']
    
    return frequency_df

# Ví dụ sử dụng hàm
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'A', 'C', 'B', 'B', 'C', 'C', 'A', 'C', 'B']
}
df = pd.DataFrame(data)

result = frequency_distribution(df, 'Category')
print(result)
```
